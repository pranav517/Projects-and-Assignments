# Step 3: Data Preparation

```python
car_pricing['fueltype']=car_pricing['fueltype'].map({"gas" : 0, "diesel" : 1})
car_pricing['aspiration']=car_pricing['aspiration'].map({'std' : 0, 'turbo' : 1})
car_pricing['doornumber']=car_pricing['doornumber'].map({'two' : 0, 'four' : 1})
car_pricing['enginelocation']=car_pricing['enginelocation'].map({'front' : 0, 'rear' : 1})
```

```python
car_pricing.head()
```

### Dummy Variables

```python
varlist=['carbody','drivewheel','enginetype','cylindernumber','fuelsystem','CarName']
```

```python
# generating dummy variables upto n-1 depth
status=pd.get_dummies(car_pricing[varlist],drop_first=True)
```

```python
status.head()
```

```python
# joining the two dataframes
car_pricing=pd.concat([car_pricing,status],axis=1)
```

```python
car_pricing.drop(varlist,axis=1,inplace=True)
```

```python
car_pricing.head()
```

```python
car_pricing.info()
```

# Step 4: Splitting data into training and testing sets

```python
# importing the necessary modules and implementing them
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test=train_test_split(car_pricing,train_size=0.7,test_size=0.3,random_state=100)
```

```python
df_train.columns
```

```python
df_test.columns
```

### Rescaling

Due to considerable variation in the scale of numerical variables,we'll consider normalisation

```python
from sklearn.preprocessing import MinMaxScaler
# instantiate
scaler=MinMaxScaler()
numerical_vars=['enginesize','curbweight','horsepower','carwidth','carlength','wheelbase','boreratio','carheight','stroke','price']
```

```python
df_train[numerical_vars]=scaler.fit_transform(df_train[numerical_vars])
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
df_train.describe()
```

```python
df_train.corr()['price'].sort_values(ascending=False)
```

Choosing engine size as the first variable

```python
y_train=df_train.pop('price')
x_train=df_train
```

# Step 5: Building a Linear Model 

```python
import statsmodels.api as sm
x_train_lm=sm.add_constant(x_train['enginesize'])
# fitting the model using the ordinary least squares
lr=sm.OLS(y_train,x_train_lm).fit()
```

```python
lr.params
```

```python
plt.scatter(x_train_lm.iloc[:,1],y_train)
plt.plot(x_train_lm.iloc[:,1],-0.072888+1.210676*x_train_lm.iloc[:,1],'r')
```

```python
print(lr.summary())
```

#### The variance explained is of about 75.3%.Taking the second most correlated another variable ,

```python
x_train_lm=x_train[['enginesize','curbweight']]
x_train_lm=sm.add_constant(x_train_lm)
lr=sm.OLS(y_train,x_train_lm).fit()
lr.params
```

```python
print(lr.summary())
```

The model has improved,as is evident from the increase in the value of adjusted R^2
