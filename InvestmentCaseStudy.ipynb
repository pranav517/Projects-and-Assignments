```python
#importing pandas and numpy libraries
import numpy as np
import pandas as pd
```

```python
#Checkpoints - Part 1
#Checkpoint 1: Data Cleaning 1

#Loading data from companies txt file
companies = pd.read_csv("Companies.txt",delimiter = "\t",encoding = 'latin1')


#Loading data from rounds2 csv file
rounds2 = pd.read_csv("rounds2.csv",encoding = 'latin1')
```

```python
#Checkpoint 4  - Sector Analysis 1
#Extracting primary_sector for each company in master_frame into a new column namely primary_sector

#Replacing all NaN values in category_list column with empty string
master_frame['category_list'].fillna("",inplace=True)

#Custom function to determine value of primary_sector
def extractSector(x):
    if x.count('|') >= 1:
        arr = x.split("|")
        if len(arr) > 0:
            return arr[0]
    else:
        return x
    
```

```python
#Apply custom function to add new primary_sector column to master_Frame for each company
master_frame['primary_sector'] = master_frame['category_list'].apply(extractSector)
```

```python
master_frame
```

```python
#Importing mapping.csv into a new mapping data frame
mapping = pd.read_csv("mapping.csv")
```

```python
#Replacing 0 with 'na' in category_list column
mapping['category_list'] = mapping['category_list'].str.replace('0','na')
```

```python
mapping
```

```python
#Custom function to determine main_sector for each primary_sector
def getMainSector(x):
    cat_list = mapping.loc[(mapping.category_list == x)]
    if len(cat_list) != 0 :
        if cat_list.iloc[0]['Automotive & Sports'] == 1 :
            return 'Automotive & Sports'
        elif cat_list.iloc[0]['Cleantech / Semiconductors'] == 1:
            return 'Cleantech / Semiconductors'
        elif cat_list.iloc[0]['Entertainment'] == 1:
            return 'Entertainment'
        elif cat_list.iloc[0]['Health'] == 1:
            return 'Health'
        elif cat_list.iloc[0]['Manufacturing'] == 1:
            return 'Manufacturing'
        elif cat_list.iloc[0]['News, Search and Messaging'] == 1:
            return 'News, Search and Messaging'
        elif cat_list.iloc[0]['Others'] == 1:
            return 'Others'
        elif cat_list.iloc[0]['Social, Finance, Analytics, Advertising'] == 1:
            return 'Social, Finance, Analytics, Advertising'
    else:
        return 'NA'
```

```python
#Creating new main_sector column in master frame to have mapped values of primary_sector
master_frame['main_sector'] = master_frame['primary_sector'].apply(getMainSector)
```
